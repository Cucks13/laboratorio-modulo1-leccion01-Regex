#### 1.1 La información de cada artista(nombre, correo de contacto, número de contacto y sitio web).

```python
lista_contenido = contenido.split("./")
lista_artista_palabras = lista_contenido[4]

patron_numero = r"-\d{1,2}\s\d{9}"
telefono = re.findall(patron_numero, lista_artista_palabras)[0]
telefono
```

```python
patron_nomnbre = r"[A-Z]\w+"
nombre = re.findall(patron_nomnbre, lista_artista_palabras)[0]
nombre
```

```python

patron_correo = r"\s(.+@.+)\s-\d"

correo = re.findall(patron_correo, lista_artista_palabras)[0]
correo

```

```python
patron_web = r"(https?://[^\s]+)"
web = re.findall(patron_web, lista_artista_palabras)[0]
web
```

```python
datos_personales = (nombre, telefono, correo, web)

print(datos_personales)
```

#### 1.2 La información de cada canción(título, duración y fecha de lanzamiento).

```python
patron_duracion = r"(\d{1,2}:\d{2})"
duracion = re.findall(patron_duracion, lista_artista_palabras)[0]
duracion
```

```python
patron_fecha = r"\d{1,2}:\d{2} (\d{4})"
fecha = re.findall(patron_fecha, lista_artista_palabras)[0]
fecha
```

```python
patron_titulo = r"https?://[^\s]+ ([\w\s]+) -"
titulo = re.findall(patron_titulo, lista_artista_palabras)[0]
titulo
```
